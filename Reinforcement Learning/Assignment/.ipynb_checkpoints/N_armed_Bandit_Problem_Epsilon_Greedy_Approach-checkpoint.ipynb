## N-armed bandit problem $\epsilon$-greedy approach
### Given, the following:
-  $n$ = 10,
-  a = 1, 2, 3, ..., 10,
-  $q(a)$ ~ $N(0,1)$ $\forall$ a,
-  at $t^{th}$ step, the actual reward is $R_{t} = q(A_{t}) + N(0,1)$, where $A_{t}$ is the action taken at the $t^{th}$ step,
-  Number of tasks = 2000,
-  Number of steps in each task = 1000.

### To find:
-  Optimal action taken for each method by averaging over bandits.
-  Average reward received by averaging over bandits.



```python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib as plt
```

### Deifine the $n$-armed bandit problem as follows:

```python
class Bandit_prob_def:
    
    def __init__(self): # initialize and define the attributes
        
        self.N_sample_action = np.zeros(10)               # number of times an action is sampled
        self.actual_arm_vals = np.random.normal(0, 1, 10) # actual arm values
        self.est_arm_vals = np.zeros(10)                  # estimated values of the arms
        
    def choose_action_value_method(self, epsilon): # choose between greedy and non-greedy action, epsilon in [0, 1].
        
        rand_num = np.random.random()              # selects a sample from an uniform distribution with a = 0, b = 1
        if epsilon>rand_num:                       # P(rand_num<epsilon) = epsilon, with probability epsilon we choose to explore
            return np.random.randint(10)
        else:
            return np.argmax(self.est_arm_vals)    # gives the index of the maximum value
    
    def action_reward(self, action): 
        reward = self.actual_arm_vals[action]+np.random.normal(0,1)
        return reward # the reward for each action
    
    def est_arm_vals_update(self, reward, action): # update the estimated values of the arms
        
        self.N_sample_action[action] += 1          # add 1 when the action is sampled
        k = self.N_sample_action[action]           # it is the number of times the action is sampled
        self.est_arm_vals[action] += (1./k)*(reward - self.est_arm_vals[action]) # using sample average method with incremental implementation for calculating the value estimates
    
    def n_optimal_actions(self, action): # method to check if the action selected is an optimal one
        
        if action == np.argmax(self.actual_arm_vals):
            return 1
        else:
            return 0
        
    
    def expt_steps(self, N_steps, epsilon): 
        reward = []
        optimal_action = []
        
        for i in range(N_steps):
            action = self.choose_action_value_method(epsilon) # choose a non-greedy action with epsilon probability
            R = self.action_reward(action)                    # get a reward for the chosen action
            opt_act = self.n_optimal_actions(action)          # check whether the action selected is optimal or not
            self.est_arm_vals_update(R, action)               # update the arm value estimates
            reward.append(R)                                  # append all reward
            optimal_action.append(opt_act)                    # append all optimal action indicators
        return np.array(reward), np.array(optimal_action)     # return rewards and optimal action indicators
```

### Perform N tasks, here N=2000

```python
epsilon = np.array([0, 0.01, 0.1])
N_steps = 1000  # number of steps 
N_tasks = 2000  # number of tasks

average_reward_epsi_0_00 = np.zeros(N_steps)
average_reward_epsi_0_01 = np.zeros(N_steps)
average_reward_epsi_0_1 = np.zeros(N_steps)

average_optimal_action_0_00 = np.zeros(N_steps)
average_optimal_action_0_01 = np.zeros(N_steps)
average_optimal_action_0_1 = np.zeros(N_steps)

for i in range(N_tasks):
    b1 = Bandit_prob_def()
    r1, o1 = b1.expt_steps(N_steps, 0)
    average_reward_epsi_0_00 += r1
    average_optimal_action_0_00 += o1
    
    b2 = Bandit_prob_def()
    r2, o2 = b2.expt_steps(N_steps, 0.01)
    average_reward_epsi_0_01 += r2
    average_optimal_action_0_01 += o2
    
    b3 = Bandit_prob_def()
    r3, o3 = b3.expt_steps(N_steps, 0.1)
    average_reward_epsi_0_1 += r3
    average_optimal_action_0_1 += o3
```

### Plot the results

```python
plt.figure(figsize=(20, 10))
plt.plot(average_reward_epsi_0_00/N_tasks,label="eps = 0.0")
plt.plot(average_reward_epsi_0_01/N_tasks,label="eps = 0.01")
plt.plot(average_reward_epsi_0_1/N_tasks,label="eps = 0.1")
plt.legend()

plt.figure(figsize=(20, 10))
plt.plot(average_optimal_action_0_00/N_tasks,label="eps = 0.0")
plt.plot(average_optimal_action_0_01/N_tasks,label="eps = 0.01")
plt.plot(average_optimal_action_0_1/N_tasks,label="eps = 0.1")
plt.legend()
```
