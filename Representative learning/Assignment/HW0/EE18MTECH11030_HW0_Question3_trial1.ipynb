```python
import numpy as np
import matplotlib.pyplot as plt
import random
import statistics
```

### Define the MLE for the distributions as follows:

```python
class MLE():
    def binomial_dist(self, X, n):
        p = sum(X)/len(X)
#         p = np.mean(X)
        return p/n
    
    def Poission_dist(self, X):
        lam = sum(X)/len(X)
        return lam
    
    def exp_dist(self, X):
        lam = 1/(sum(X)/len(X))
        return 1/lam
    
    def Gaussian_dist(self, X):
        mean = sum(X)/len(X)
        var = (1/len(X))*(sum(np.square(X-mean)))
        return mean, var, np.sqrt(var)
    
    def Laplacian_dist(self, X):
        mean = statistics.median(X)
        b = (1/len(X))*sum(abs(X-mean))
        return mean, b
```

### Verify the results

#### 1) Binomial Distribution

```python
n = 100
N = 100000
p_org = 0.65
X_org =  np.random.binomial(n, p_org, N)
# X_org = (X_org-min(X_org))/(max(X_org)-min(X_org))

obj = MLE()
p_est = obj.binomial_dist(X_org, n)
print('p_est = ', p_est)
X_est = np.random.binomial(n, p_est, N)
# X_est = (X_est-min(X_est))/(max(X_est)-min(X_est))

_,_,_ = plt.hist(X_org, bins = 10)
plt.show()

_,_,_ = plt.hist(X_est, bins = 10)
plt.show()
```

#### 2) Poission Distribution

```python
N = 100000
lam_org = 5
X_org =  np.random.poisson(lam_org, N)
# X_org = (X_org-min(X_org))/(max(X_org)-min(X_org))

obj = MLE()
lam_est = obj.Poission_dist(X_org)
print('lam_est = ', lam_est)

X_est = np.random.poisson(lam_est, N)
# X_est = (X_est-min(X_est))/(max(X_est)-min(X_est))

_,_,_ = plt.hist(X_org, bins = 10)
plt.show()

_,_,_ = plt.hist(X_est, bins = 10)
plt.show()
```

#### 3) Exponential Distribution

```python
N = 100000
lam_org = 5
X_org =  np.random.exponential(lam_org, N)
# X_org = (X_org-min(X_org))/(max(X_org)-min(X_org))

obj = MLE()
lam_est = obj.exp_dist(X_org)
print('lam_est = ', lam_est)

X_est = np.random.exponential(lam_est, N)
# X_est = (X_est-min(X_est))/(max(X_est)-min(X_est))

_,_,_ = plt.hist(X_org, bins = 10)
plt.show()

_,_,_ = plt.hist(X_est, bins = 10)
plt.show()
```

#### 4) Gaussian Distribution

```python
N = 100000
mean_org = 5
std_org = 0.8
X_org =  np.random.normal(mean_org, std_org, N)
# X_org = (X_org-min(X_org))/(max(X_org)-min(X_org))

obj = MLE()
mean_est, var_est, std_est = obj.Gaussian_dist(X_org)
print('mean_est, std_est = ', mean_est, ',' ,std_est)

X_est = np.random.normal(mean_est, std_est, N)
# X_est = (X_est-min(X_est))/(max(X_est)-min(X_est))

_,_,_ = plt.hist(X_org, bins = 10)
plt.show()

_,_,_ = plt.hist(X_est, bins = 10)
plt.show()
```

#### 5) Laplacian Distribution

```python
N = 100000
mu_org = 5
b_org = 0.8
X_org =  np.random.laplace(mu_org, b_org, N)
# X_org = (X_org-min(X_org))/(max(X_org)-min(X_org))

obj = MLE()
mu_est, b_est = obj.Laplacian_dist(X_org)
print('mu_est, b_est = ', mean_est, ',' ,std_est)

X_est = np.random.laplace(mu_est, b_est, N)
# X_est = (X_est-min(X_est))/(max(X_est)-min(X_est))

_,_,_ = plt.hist(X_org, bins = 10)
plt.show()

_,_,_ = plt.hist(X_est, bins = 10)
plt.show()
```
